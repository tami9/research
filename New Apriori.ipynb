```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('Sample data.csv', header=None)
dataset
```

```python
len(dataset.columns)
```

```python
transactions = []
for i in range(0, 100):
    transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
transactions
```

```python
items = []
for i in range(0, len(transactions)):
    items.extend(transactions[i])
items
```

```python
uniqueItems = list(set(items))
uniqueItems.remove('nan')
uniqueItems
```

```python
pair = []
for j in range(0, len(uniqueItems)):
    k = 1
    while k <= len(uniqueItems):
        try:
            pair.append([uniqueItems[j], uniqueItems[j+k]])
        except IndexError:
            pass
        k = k + 1
pair
```

```python
scores = []
for i in pair:
    condition = []
    for item in i:
        condition.append('("%s") in s' %item)
    create_code = ('[s for s in transactions if ' + ' and '.join(condition) + ']')
    scores.append(len(eval(create_code))/len(dataset))
scores
```

```python
rank = pd.DataFrame([pair, scores]).T
rank.columns = ['Pair', 'Score']
rank.sort_values('Score', ascending=False)
```

```python
score = []
confidence = []
for i in duo:
    cond = []
    for item in i:
        cond.append('("%s") in s' %item)
    mycode = ('[s for s in transactions if ' + ' and '.join(cond) + ']')
    confidence_code = ('[x for x in transactions if ("%s") in x]' %i[0])
    print(confidence_code)
    #mycode = "print 'hello world'"
    score.append(len(eval(mycode))/len(dataset))
    confidence.append(len(eval(mycode))/len(eval(confidence_code)))
```

```python
confidence
```

```python
ranking = pd.DataFrame([duo, score, confidence]).T
ranking.columns = ['Pairs', 'Score', 'Confidence']
ranking.sort_values('Score', ascending=False).head(10)
```

```python
ranking[ranking.Score > 0.05].sort_values('Score', ascending=False)
```
